## 1. Introducción

El presente informe tiene como propósito asesorar al **Sr. Juan** sobre cuál de las cuatro tiendas de su portafolio comercial presenta el menor potencial de rentabilidad y desempeño global, con el fin de identificar la unidad de negocio que debería **vender**, permitiéndole concentrar sus recursos en las tiendas más sólidas y competitivas.

El análisis se realizó sobre un conjunto de aproximadamente **9.435 transacciones** distribuidas entre las cuatro tiendas, evaluando cinco dimensiones clave:

-  Ingresos totales generados por tienda
-  Categorías de productos más y menos vendidas
-  Calificaciones promedio otorgadas por los clientes
-  Productos con mayor y menor rotación
-  Costo promedio de envío por tienda

## 2. Desarrollo del Análisis

Esta brecha no es menor: representa aproximadamente el 10% del ingreso total de la tienda líder.

### 2.3  Calificaciones Promedio de los Clientes

La satisfacción del cliente es un predictor de fidelización y recompra a largo plazo.

**Hallazgo:** Las diferencias son pequeñas (menos de 0.1 puntos), pero es importante destacar que la **Tienda 1**, siendo la de mayor ingreso, tiene el **rating más bajo** (3.98). La Tienda 4, a pesar de sus bajos ingresos, logra mantener una calificación aceptable de 4.00, lo que sugiere que su problema es de **volumen y precio, no de calidad de servicio**.

## 4. Conclusión y Recomendación Final

###  Recomendación: El Sr. Juan debe vender la **Tienda 4**

Tras el análisis exhaustivo de las cinco dimensiones evaluadas, la evidencia apunta de manera consistente y convergente hacia la **Tienda 4** como la unidad de negocio con menor potencial relativo dentro del portafolio.

**Las razones que respaldan esta decisión son:**

1. **Menor ingreso total del portafolio:** Con $1.038.375.700, la Tienda 4 es la única que se aleja significativamente del resto, con una diferencia de más de $112 millones respecto a la Tienda 1. Esta brecha estructural es difícil de revertir sin una reestructuración profunda del modelo de negocio.

2. **Menor número de transacciones registradas:** Con 2.358 registros frente a los 2.359 de las demás tiendas, aunque la diferencia es mínima en términos absolutos, confirma una tendencia de menor actividad comercial.

3. **Debilidad en categorías estratégicas:** La Tienda 4 tiene el peor desempeño en Electrodomésticos (254 ventas) e Instrumentos musicales (170 ventas), categorías que en otras tiendas generan mayor tracción y margen.

4. **La ventaja logística no genera valor:** Su costo de envío más bajo ($23.459) es la única métrica donde lidera, pero esta eficiencia operativa no se está convirtiendo en volumen de ventas ni en satisfacción superior. Representa una fortaleza desperdiciada.

5. **Satisfacción del cliente aceptable pero no diferencial:** Un rating de 4.00 es correcto, pero no compensa las brechas en ingresos y volumen frente a Tiendas 2 y 3, que tienen mejor calificación y mayores ingresos simultáneamente.

###  Consideración Estratégica Final

Las **Tiendas 1, 2 y 3** presentan perfiles más equilibrados y complementarios entre sí. La Tienda 1 es la más rentable; la Tienda 3 tiene la mejor satisfacción del cliente; la Tienda 2 tiene el producto más vendido del sistema. Vender la Tienda 4 le permitirá al Sr. Juan **liberar capital**, **reducir carga operativa** y **reinvertir en las tres tiendas restantes**, que demuestran mayor solidez y potencial de crecimiento sostenido.
